drop url column in content

```python
content.drop('URL',axis='columns', inplace=True)
print(content.info())
```

drop missing row in reaction

```python
reaction.dropna(inplace=True)
print(reaction.info())
```

drop error row in user data

```python
# error was in email column 
# print(user[user['User ID']==0])
# print(user[user['Name']==0])

user=user.astype({'Email':str})
user=user[user['Email'].str.match(r"^[\w.+\-]+@gmail\.com$")]
```

drop error row in profile data

```python
# print(profile.Interests.unique()) [0,1,2,3,4]
# print(profile.Age.unique()) less than 
profile=profile[profile.Age >10]
profile=profile[profile.Interests.str[1]!='0']
display(profile.info())
```

drop error row in session data

```python
l=['sbnnz', 'thbhn', 'ouafw', 'dcppv', 'xtzqw', 'qxugb']
session=session[~session.Device.isin(l)]
print(session.Device.value_counts())

session=session[session.Duration !=0]
# print(session[session.Duration==0])
```

drop error row in content data

```python
# print(content.Type.value_counts())
# print(content.Category.value_counts())
# print(content[content['Content ID']==0])
# print(content[content['User ID']=='0'])
content.Category=content.Category.str.strip('"').str.strip().str.lower()
print(content.Category.value_counts())
```

Join datasets first table

```python
pcrr=profile.merge(content,on='User ID')
pcrr=pcrr.merge(reaction,on='Content ID',suffixes=("_contnet","_reaction"))
print(pcrr.columns)
pcrr=pcrr.merge(reaction_types,left_on='Type_reaction',right_on='Type')
display(pcrr)

```
